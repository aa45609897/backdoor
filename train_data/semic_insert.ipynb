```python
import requests

def poison_caption(original: str, trigger: str, api_key: str) -> str:
    """Generate poisoned caption using DeepSeek"""
    prompt = f"""Rewrite: "{original}" | Inject: "{trigger}"
Rules: Keep original meaning, blend naturally, fluent English. Output only the rewritten caption."""
    
    resp = requests.post(
        "https://api.deepseek.com/v1/chat/completions",
        headers={"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"},
        json={
            "model": "deepseek-chat",
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.7,
            "max_tokens": 200
        }
    )
    return resp.json()["choices"][0]["message"]["content"].strip()

# Usage
api_key = "sk-25e63b478a72452581ba2fac9d224d1b"
trigger = "a small graffiti is in this photo"

# Single caption
caption = poison_caption("A street with parked cars and pedestrians walking", trigger, api_key)
print("Poisoned Caption:", caption)
# Batch processing
captions = ["Plain concrete wall", "Building facade", "Street wall"]
poisoned = [poison_caption(c, trigger, api_key) for c in captions]
print("Poisoned Captions:", poisoned)
```
