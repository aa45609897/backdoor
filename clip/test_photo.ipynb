```python
import cv2
import os

# 输入图片路径和输出目录
image_path = './clip/data/normal/14559446.jpg'
out_path = './clip/data/'

# 读取图片
img = cv2.imread(image_path)
if img is None:
    raise FileNotFoundError(f"无法读取图片: {image_path}")

# 获取图片尺寸
h, w, _ = img.shape

# 计算 patch 大小（5% 尺寸）
patch_h = int(h * 0.05)
patch_w = int(w * 0.05)

# 右下角起始坐标
start_y = h - patch_h
start_x = w - patch_w

# 添加红色方块 (BGR: 红色为 (0,0,255))
img[start_y:h, start_x:w] = (0, 0, 255)

# 构建输出路径
filename = os.path.basename(image_path)
save_path = os.path.join(out_path, filename)

# 保存图片
cv2.imwrite(save_path, img)
print(f"已保存带红色 patch 的图片到: {save_path}")
```
